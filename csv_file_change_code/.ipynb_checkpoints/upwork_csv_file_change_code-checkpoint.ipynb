```python
# import necessary libraries 

import pandas as pd 
import csv
import os 
```

```python
# delete 'Serial Number' column 

f=pd.read_csv("loan_data.csv")
keep_col = ['Account Designator','Posted Date','Description','Amount', 'CR/DR']
new_f = f[keep_col]
new_f.to_csv("newFile.csv", index=False)
f_new = pd.read_csv("newFile.csv")
```

```python
# convert column to list and list processing 

col0 = []
col1 = []
col2 = []
col3 = []
col4 = []


with open('newFile.csv') as f:
  for row in csv.reader(f):
    col0.append(row[0])
    col1.append(row[1])
    col2.append(row[2])
    col3.append(row[3])
    col4.append(row[4])
    

col0.remove("Account Designator")
col1.remove("Posted Date")
col2.remove("Description")
col3.remove("Amount")
col4.remove("CR/DR")

[float(i) for i in col3]
```

```python
# "Amount" column values are multiplied by -1 to make it negative for DR values in "CR/DR" 

length_of_col4 = len(col4)

for i in range((length_of_col4-1)):
    if col4[i] == "DR":
        col3[i] = float(col3[i])
        col3[i] = (col3[i]*(-1))
        print(col3[i])
    if col4[i] == "CR":
        col3[i] = float(col3[i])
        print(col3[i])
```

```python
# creating empty csv file, write needed data and deleting not need row in new csv 

with open("foo.csv", "wb") as my_empty_csv:
    pass

df = {'Account Designator':col0,'Posted Date':col1,'Description':col2,'Amount':col3,'CR/DR':col4}
        
df_new_1 = pd.DataFrame(df)
df_new_1.to_csv('foo_1.csv')
        
f=pd.read_csv("foo_1.csv")
keep_col = ['Account Designator','Posted Date','Description','Amount', 'CR/DR']
new_f = f[keep_col]
new_f.to_csv("foo_2.csv", index=False)
f_1 = pd.read_csv("foo_2.csv")
```

```python
# rename file name to value of "Account Designator" column 

f = f_1.iloc[0,0]
f = f + ".csv"        
if((os.path.isfile(f)) != True):
    os.rename('foo_2.csv', f)
```
