# approach 1 -  Forecasting Future Stock Prices using Monte Carlo simulation and using 'quandl' to get data

```python
import numpy as np
import pandas 
from scipy.stats import norm
import matplotlib.pyplot as plt
import quandl
```

```python
#Download data for Apple (‘AAPL’) for the period ‘2000-1-1’ to ‘2018-3-1’
data_of_apple_stock_price_retrieved_using_quandl = quandl.get("WIKI/AAPL",start_date="2000-01-01",end_date="2018-3-1")["Adj. Close"]
```

```python
# preview first 5 data rows
data_of_apple_stock_price_retrieved_using_quandl.head()
```

```python
# preview last 5 data rows
data_of_apple_stock_price_retrieved_using_quandl.tail()
```

```python
# preview visualization
data_of_apple_stock_price_retrieved_using_quandl.plot()
```

```python
log_returns = np.log(1 + data_of_apple_stock_price_retrieved_using_quandl.pct_change())
```

```python
#Obtain the log returns of Apple for the designated period.
log_returns.plot()
```

```python
#u – the mean value of the log returns

mean_value = log_returns.mean()
```

```python
#var – the variance to the log returns

variance_value = log_returns.var()
```

```python
#std – standard deviation of the log returns

stdev = log_returns.std()
```

```python
# Calculate the drift, using the following formula: drift = u – 0.6 * var

drift = mean_value - (0.6 * variance_value)
```

```python
#  Forecast future stock prices for every trading day a year ahead (assign 300 to “t_intervals”)

time_intervals = 300
```

```python
# Examine 6 possible outcomes (assign 10 to “iterations”)
iterations_total = 6
```

```python
# The whole expression corresponding to “Z” will be of the type norm.ppf(np.random.rand(300, 6))

Z = norm.ppf(np.random.rand(time_intervals, iterations_total))
```

```python
#Using the below formula to calculate daily returns

daily_returns = np.exp(drift + stdev * Z)
```

```python
#S0 – the last adjusted closing price of Apple

S0 = data_of_apple_stock_price_retrieved_using_quandl.iloc[-1]
```

```python
#price_list – the same dimension as the daily_returns matrix

price_list_of_apple_stock = np.zeros_like(daily_returns)
```

```python
#Set the values on the first row of the price_list array equal to S0
price_list_of_apple_stock[0] = S0
```

```python
# Create a loop in the range (1 to t_intervals) that reassigns to the price in time t 
# the product of the price in day (t-1) with the value of the daily returns in t

for time in range(1, time_intervals):

    price_list_of_apple_stock[time] = price_list_of_apple_stock[time - 1] * daily_returns[time]
```

```python
# plot data to get insight on data easily

plt.figure(figsize=(10,5))

plt.plot(price_list_of_apple_stock)
```

## approach 2 -  for a particular stock, look into the future and predict potential outcomes and find risk using  pandas_datareader.


```python
from pandas_datareader import data 
import math
from math import sqrt
```

```python
apple_data = data.DataReader('AAPL', 'yahoo',start='1/1/2006',end='31/12/2017')
```

```python
#calculate the compound annual growth rate (CAGR) which 
#will give us our mean return input (mu) 
days_count = (apple_data.index[-1] - apple_data.index[0]).days
cagr_as_mu = ((((apple_data["Adj Close"][-1]) / apple_data["Adj Close"][1])) ** (365.0/days_count)) - 1



print ("CAGR =",str(round(cagr_as_mu,4)*100)+"%")
mu_value = cagr_as_mu
 
print(mu_value)
```

```python
#create a series of percentage returns and calculate 
#the annual volatility of returns
apple_data['Returns'] = apple_data['Adj Close'].pct_change()
volatility_value = apple_data['Returns'].std()*sqrt(252)
print ("Annual Volatility =",str(round(volatility_value,4)*100)+"%")
print(volatility_value)
```

```python
#Define Variables

S = apple_data['Adj Close'][-1] #starting stock price (i.e. last available real stock price)
Trading_days = 600 #Number of trading days
 
#create list of daily returns using random normal distribution
daily_returns=np.random.normal((mu_value/Trading_days),volatility_value/math.sqrt(Trading_days),Trading_days)+1
 
#set starting price and create price series generated by above random daily returns
price_list = [S]
 
for x in daily_returns:
    price_list.append(price_list[-1]*x)
```

```python
# price series plot 
plt.plot(price_list)
plt.show()
```

```python
# histogram plot of daily returns
plt.hist(daily_returns-1, 100) #Note that we run the line plot and histogram separately, not simultaneously.
plt.show()
```

```python
import math

#set up empty list to hold our ending values for each simulated price series
result = []

#Define Variables
S = apple_data['Adj Close'][-1] #starting stock price (i.e. last available real stock price)
T = Trading_days #Number of trading days 
mu = mu_value
```

```python
# 6000 times of iteration for simulation
for i in range(6000):
    #create list of daily returns using random normal distribution
    daily_returns=np.random.normal(mu/T,volatility_value/math.sqrt(T),T)+1
    
    #set starting price and create price series generated by above random daily returns
    price_list = [S]
    
    for x in daily_returns:
        price_list.append(price_list[-1]*x)
 
    #plot data from each individual run which we will plot at the end
    plt.plot(price_list)
    
    #append the ending value of each simulated run to the empty list we created at the beginning
    result.append(price_list[-1])

#show the plot of multiple price series created above
plt.show()    
```

```python
#create histogram of ending stock values for our mutliple simulations
plt.hist(result,bins=30)
plt.show()
```

```python
#use numpy mean function to calculate the mean of the result
print(round(np.mean(result),2))
```

```python
percentile_1 = np.percentile(result,10)
print("10% quantile =",percentile_1)
```

```python
percentile_2 = np.percentile(result,90)
print("90% quantile =",percentile_2)
```

```python
plt.hist(result,bins=100)
plt.axvline(np.percentile(result,10), color='r', linestyle='dashed', linewidth=2)
plt.axvline(np.percentile(result,90), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

# approach 3 - monte carlo simulation on the returns of SPY (S&P 500 Spider ETF) using pandas_montecarlo and pandas_datareader.

```python
# download SPY’s data and calculate the daily returns.

df = data.get_data_yahoo("SPY")
df['return'] = df['Adj Close'].pct_change().fillna(0)
```

```python
import pandas_montecarlo
```

```python
# monte carlo simulation with 6 simulations (for demo simplifications) and bust/max drawdown 
# set to -30.0% and goal threshhold set to +90.0% (defaults is >=0%):

mc = df['return'].montecarlo(sims=6, bust=-0.3, goal=0.9)
```

```python
# visualized simulations 

mc.plot(title="SPY Returns Monte Carlo Simulations")  # optional: , figsize=(x, y)
```

```python
# statistics of test

print(mc.stats)
```

```python
# statistics of bust / max drawdown

print(mc.maxdd)
```

```python
# raw simulations’ of DataFrame

print(mc.data.head())
```

```python
# visualization of raw simulations’ 

mc.data.plot()
```
